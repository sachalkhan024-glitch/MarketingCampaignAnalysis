```python
customers=df[[ 'ID', 'Year_Birth', 'Education', 'Marital_Status', 'Income', 'Kidhome', 'Teenhome', 'Dt_Customer', 'Recency', 'Country', 'Complain']]
```

```python
product_spending=df[[ 'ID', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']]
```

```python
purchase_channels=df[[ 'ID', 'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']]
```

```python
campaign_responses=df[['ID', 'AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response']]
```

```python
#dividing tables to load in sql
```

```python
import sqlite3
```

```python
con=sqlite3.connect('freshchoice.db')
```

```python
customers.to_sql('customers', con, if_exists='replace', index=False)
product_spending.to_sql('product_spending', con, if_exists='replace', index=False)
purchase_channels.to_sql('purchase_channels', con, if_exists='replace', index=False)
campaign_responses.to_sql('campaign_responses', con, if_exists='replace', index=False)
```

```python
pd.read_sql("Select * from sqlite_master Where type= 'table'",con)
```

```python
pd.read_sql('''SELECT Response, ROUND(AVG(Income), 0) AS avg_income
                FROM customers
                JOIN campaign_responses 
                ON customers.ID = campaign_responses.ID
                WHERE Income IS NOT NULL
                GROUP BY Response;''', con)
```

```python
pd.read_sql('''SELECT Education, COUNT(*) AS count, SUM(Response) AS ACCEPTANCE
                FROM customers
                JOIN campaign_responses
                ON customers.ID=campaign_responses.ID
                GROUP BY Education''', con)
```

```python
pd.read_sql('''SELECT Kidhome, SUM(cr.Response) AS accepted, COUNT(*) AS total
FROM customers c
JOIN campaign_responses cr ON c.ID = cr.ID
GROUP BY Kidhome;''', con)
```

```python
pd.read_sql('''SELECT Kidhome, Teenhome, SUM(cr.Response) AS accepted, COUNT(*) AS total
FROM customers c
JOIN campaign_responses cr ON c.ID = cr.ID
GROUP BY Kidhome, Teenhome;''',con)
```

```python
pd.read_sql('''SELECT 
  CASE WHEN Income >= 80000 THEN 'High' ELSE 'Under 80k' END AS income_group,
  CASE WHEN Kidhome = 0 AND Teenhome = 0 THEN 'No Kids' ELSE 'Has Kids/Teens' END AS family_group,
  SUM(cr.Response) AS accepted,
  COUNT(*) AS total
FROM customers c
JOIN campaign_responses cr ON c.ID = cr.ID
GROUP BY income_group, family_group
ORDER BY accepted DESC;''',con)
```

```python
#acceptance by country
pd.read_sql('''SELECT Country, COUNT(*) AS total, SUM(cr.Response) AS accepted
FROM customers c
JOIN campaign_responses cr ON c.ID = cr.ID
GROUP BY Country
ORDER BY accepted DESC;''', con)
```

```python
#average product spending by acceptance status
pd.read_sql('''SELECT cr.Response,
  ROUND(AVG(ps.MntWines), 0) AS avg_wines,
  ROUND(AVG(ps.MntFruits), 0) AS avg_fruits,
  ROUND(AVG(ps.MntMeatProducts), 0) AS avg_meat,
  ROUND(AVG(ps.MntFishProducts), 0) AS avg_fish,
  ROUND(AVG(ps.MntSweetProducts), 0) AS avg_sweets,
  ROUND(AVG(ps.MntGoldProds), 0) AS avg_gold
FROM campaign_responses cr
JOIN product_spending ps ON cr.ID = ps.ID
GROUP BY cr.Response;''', con)
```

```python
#average purchase channel usage by acceptance status
pd.read_sql('''SELECT cr.Response,
  ROUND(AVG(pc.NumDealsPurchases), 1) AS avg_deals,
  ROUND(AVG(pc.NumWebPurchases), 1) AS avg_web,
  ROUND(AVG(pc.NumCatalogPurchases), 1) AS avg_catalog,
  ROUND(AVG(pc.NumStorePurchases), 1) AS avg_store,
  ROUND(AVG(pc.NumWebVisitsMonth), 1) AS avg_web_visits
FROM campaign_responses cr
JOIN purchase_channels pc ON cr.ID = pc.ID
GROUP BY cr.Response;''', con)
```

```python
#average previous campaign acceptance by campaign 6 response
pd.read_sql('''SELECT cr.Response,
  ROUND(AVG(cr2.AcceptedCmp1), 3) AS avg_cmp1,
  ROUND(AVG(cr2.AcceptedCmp2), 3) AS avg_cmp2,
  ROUND(AVG(cr2.AcceptedCmp3), 3) AS avg_cmp3,
  ROUND(AVG(cr2.AcceptedCmp4), 3) AS avg_cmp4,
  ROUND(AVG(cr2.AcceptedCmp5), 3) AS avg_cmp5
FROM campaign_responses cr
JOIN campaign_responses cr2 ON cr.ID = cr2.ID
GROUP BY cr.Response;''', con)
```

## Analysis Summary
After cleaning and normalizing the data, SQL queries revealed some intresting insights for FreshChoice's Campaign 6 targeting:

1. **Income**: High-income customers (>80k) accept at 40% vs 12% for lower income brackets, showing income is a big factor
2. **Family Status**: Customers with no kids/teens accept at 26.5% compared to 10.3% for those with children - big difference
3. **Combined**: High-income + no kids has a 41.5% acceptance rate, thats nearly 3x the baseline of 14.9%
4. **Education**: PhD holders accept at 20.8% while Basic education only 3.7%
5. **Country**: Acceptance rates vary by region so geographic targeting could help
6. **Product Spend**: Accepters spend way more across all products specially wine and meat
7. **Channel**: Accepters prefer catalog and store purchases, non-accepters browse the web more but dont buy

**Recommendation**: Focus Campaign 6 on high-income customers without children - this 188-person segment gave 78 acceptances at 41.5% vs the overall 14.9% which is way better.
